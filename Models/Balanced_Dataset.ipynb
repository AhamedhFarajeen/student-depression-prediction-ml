```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/AIML_Evaluation/Data/student_depression_dataset.csv")
```

```python
#STEP 3: Separate Features (X) and Target (y)
X = df.drop("Depression", axis=1)
y = df["Depression"]
```

```python
#STEP 4: Apply SMOTE to Balance the Dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
#STEP 5: Check the New Class Distribution
print("Before Balancing:")
print(y.value_counts())

print("\nAfter SMOTE Balancing:")
print(y_resampled.value_counts())
```

```python
# STEP 6: Combine and Save the Balanced Dataset
balanced_df = pd.concat(
    [pd.DataFrame(X_resampled, columns=X.columns),
     pd.DataFrame(y_resampled, columns=["Depression"])],
    axis=1
)
```

```python
# Save the new balanced dataset to your Drive
save_path = "/content/drive/MyDrive/AIML_Evaluation/Data/student_depression_balanced.csv"
balanced_df.to_csv(save_path, index=False)
print(f"\nBalanced dataset saved successfully at: {save_path}")
```

```python
#STEP 7: Split the Balanced Data into Train/Test Sets
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

print("\nData successfully split into training and testing sets!")
print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```
